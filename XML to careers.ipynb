```python
import gzip
from lxml import etree, html
```

```python
# open the gzipped wiki dump avalable here:
#  http://wfrp1e.wikia.com/wiki/Special:Statistics -> Current pages link
gzw = gzip.open("wfrp1e_pages_current.xml.gz", "r")
# parse it with lxml
wiki = etree.parse(gzw)
```

```python
#TODO: Refactor get_page_titles and get_page_content_for to share the code
def get_page_titles(wiki):
    """This helper function returns all content page titles
    from the wiki etree data. That is, all the pages with where 
    namespace tag "ns" is 0"""

    page_titles = []
    for page in wiki.getroot():
        content_page_element = False
        current_page_text = None
        current_page_title = None
        for element in page:
            tagname = element.tag.split("}")[1]
            if tagname=="title":
                current_page_title = element.text
            if tagname=="revision":
                for rev_el in element:
                    rev_el_tagname = rev_el.tag.split("}")[1]
                    if rev_el_tagname =="text":
                        text_element = rev_el
                current_page_text = text_element.text
            if tagname=="ns" and element.text == "0":
                content_page_element = True
                
        if content_page_element:
            page_titles.append( current_page_title )
    return page_titles
```

```python
def get_page_content_for(wiki, page_name):
    """Get a specific page content from the wiki etree data"""
    
    for page in wiki.getroot():
        content_page_element = False
        current_page_text = None
        current_page_title = None
        for element in page:
            tagname = element.tag.split("}")[1]
            if tagname=="title":
                current_page_title = element.text
            if tagname=="revision":
                for rev_el in element:
                    rev_el_tagname = rev_el.tag.split("}")[1]
                    if rev_el_tagname =="text":
                        text_element = rev_el
                current_page_text = text_element.text
            if tagname=="ns" and element.text == "0":
                content_page_element = True
                
        if content_page_element:
            if current_page_title.strip()==page_name:
                return current_page_text
        
    
```

```python
#TODO: Write validation of the parsed data
class Career:
    """ A WFRP 1ed career class. Fill it with parse_career or with constructor.
    """
    attr_targets = ["M","WS","BS","S","T","W","I","A","Dex","Ld","Int","Cl","WP","Fel"]
    
    def __init__(self, career_name, career_text_for_parsing=None):
        #list of strings that are names of careers
        self.name = career_name
        self.exits = []
        self.trappings = []
        self.skills = []
        self.advances = {}
        
        if career_text_for_parsing:
            self.parse_career(career_text_for_parsing)
            
        
        
    def parse_career(self, career_text):    
        state = "seek" # "careers", "skills", "trappings"
        
        for line in  career_text.split("\n"):
            lile = line.strip()
            
            # state changes
            if r"'''Career Exits'''" in line:
                state = "careers"
                continue
            elif r"'''Trappings'''" in line:
                state = "trappings"
                continue
            elif r"'''Skills'''" in line:
                state = "skills"
                continue
            
            if state=="careers":
                #print "read careers from line", line
                if len(line)>0 and line[0]=="*":
                    exit_name = line.replace("*[[","").replace("]]","").strip()
                    self.exits.append(exit_name)
                else:
                    state = "seek"
            if state=="skills":
                #print "read skills from line", line
                if len(line)>0 and line[0]=="*":
                    skill_name = line.replace("*[[","").replace("]]","").strip()
                    self.skills.append(skill_name)
                else:
                    state = "seek"        
        
            # todo: parse trappings

        
        # parse the tables from the text
        attr_inc = 0
        for cell in page_text.split("|"):
            if "<p" in cell:
                cell_tree = html.fragment_fromstring(cell)
                # the first tag within <p> should be
                #  a <span> with the cell data.
                # WARNING: this is not verified, as the format
                #  is verified below
                cell_text = cell_tree[0].text.strip()
                
                # ignore the table header
                if cell_text=="Advance Scheme":
                    pass
                # reading the attribute advances
                elif attr_inc>=len(self.attr_targets):
                    number_text = cell_text
                    if number_text=="":
                        number_text = "0"
                    attr = self.attr_targets[attr_inc-len(self.attr_targets)]
                    self.advances[attr] = int(number_text)
                    attr_inc+=1
                # verifying order of the attributes
                elif cell_text==self.attr_targets[attr_inc]:
                    attr_inc+=1
        if attr_inc!=len(self.attr_targets)*2:
            raise ValueError("the text does not describe a standard advance scheme table")
        
        
    def __str__(self):
        advances_str = []
        for attr in self.attr_targets:
            advances_str.append(attr+":+"+str(self.advances[attr]))
        return self.name+"->"+", ".join(self.exits) + "\n" + ", ".join(advances_str)
            
    
```

```python
# Parse lists
def get_careers(list_name):
    careers = []
    for line in get_page_content_for(list_name).split("\n"):
        if len(line)>0 and line[0]=="*":
            career_name = line.replace("*[[","").replace("]]","").strip()
            careers.append(career_name)
    return careers
      
basic_careers = get_careers("List Of Basic Careers")
adv_careers = get_careers("List Of Advanced Careers")

career_name = basic_careers[0]
page_text = get_page_content_for(career_name)

career = Career(career_name)
career.parse_career(page_text)
print career

        
```
